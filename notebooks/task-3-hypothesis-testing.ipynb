```python
# Create A/B groups
df_clean = df.copy()

df_clean["RiskGroup"] = df_clean["TotalClaims"].apply(lambda x: "A_LowRisk" if x == 0 else "B_HighRisk")

# Basic summary
df_clean["RiskGroup"].value_counts()
```

```python
import numpy as np

group_A = df_clean[df_clean["RiskGroup"] == "A_LowRisk"]["TotalPremium"].dropna()
group_B = df_clean[df_clean["RiskGroup"] == "B_HighRisk"]["TotalPremium"].dropna()

len(group_A), len(group_B)
```

```python
from scipy.stats import ttest_ind

t_stat, p_value = ttest_ind(group_A, group_B, equal_var=False)

t_stat, p_value
```

```python
alpha = 0.05

if p_value < alpha:
    print("Reject the null hypothesis — Groups A and B have significantly different premium behaviors.")
else:
    print("Fail to reject the null — Premiums are not significantly different between groups.")
```

```python
# Create Claim Frequency (1 if TotalClaims > 0 else 0)
df['ClaimFrequency'] = df['TotalClaims'].apply(lambda x: 1 if x > 0 else 0)

# Create Claim Severity (average amount given a claim occurred)
# Avoid division by zero by using np.where
import numpy as np
df['ClaimSeverity'] = np.where(df['TotalClaims'] > 0, df['TotalClaims'], np.nan)

# Create Margin
df['Margin'] = df['TotalPremium'] - df['TotalClaims']

# Quick check
df[['TotalClaims','ClaimFrequency','ClaimSeverity','Margin']].head()
```

```python
from scipy.stats import ttest_ind, chi2_contingency
import numpy as np

# Prepare data for t-test: pairwise comparison between provinces
provinces = df['Province'].unique()
results_province = []

for i in range(len(provinces)):
    for j in range(i+1, len(provinces)):
        prov1 = df[df['Province']==provinces[i]]['ClaimFrequency']
        prov2 = df[df['Province']==provinces[j]]['ClaimFrequency']
        t_stat, p_val = ttest_ind(prov1, prov2, equal_var=False, nan_policy='omit')
        results_province.append({
            'Province1': provinces[i],
            'Province2': provinces[j],
            't_stat': t_stat,
            'p_value': p_val
        })

# Convert to DataFrame for readability
results_province_df = pd.DataFrame(results_province)
results_province_df.sort_values('p_value').head(10)
```

```python
# Pairwise t-tests for Margin by province
results_margin = []

for i in range(len(provinces)):
    for j in range(i+1, len(provinces)):
        margin1 = df[df['Province']==provinces[i]]['Margin']
        margin2 = df[df['Province']==provinces[j]]['Margin']
        t_stat, p_val = ttest_ind(margin1, margin2, equal_var=False, nan_policy='omit')
        results_margin.append({
            'Province1': provinces[i],
            'Province2': provinces[j],
            't_stat': t_stat,
            'p_value': p_val
        })

results_margin_df = pd.DataFrame(results_margin)
results_margin_df.sort_values('p_value').head(10)
```

```python
# t-test for Claim Frequency between Women and Men
female = df[df['Gender']=='Female']['ClaimFrequency']
male = df[df['Gender']=='Male']['ClaimFrequency']

t_stat_gender, p_val_gender = ttest_ind(female, male, equal_var=False, nan_policy='omit')

print(f"T-statistic: {t_stat_gender:.3f}, P-value: {p_val_gender:.5f}")
if p_val_gender < 0.05:
    print("Reject the null hypothesis: Significant difference in risk between genders")
else:
    print("Fail to reject the null hypothesis: No significant difference in risk between genders")
```

```python
# Statistical Testing by Zip Code

import pandas as pd
from scipy.stats import ttest_ind

# Create Claim Frequency and Margin if not already done
df['ClaimFrequency'] = (df['TotalClaims'] > 0).astype(int)
df['Margin'] = df['TotalPremium'] - df['TotalClaims']

# Select top 5 zip codes by number of policies to make the test manageable
top_zipcodes = df['PostalCode'].value_counts().nlargest(5).index.tolist()

# Prepare a DataFrame to store results
zip_results = []

for i in range(len(top_zipcodes)):
    for j in range(i+1, len(top_zipcodes)):
        zip_a = df[df['PostalCode'] == top_zipcodes[i]]
        zip_b = df[df['PostalCode'] == top_zipcodes[j]]

        # T-test for Claim Frequency
        t_freq, p_freq = ttest_ind(zip_a['ClaimFrequency'], zip_b['ClaimFrequency'], equal_var=False)
        
        # T-test for Margin
        t_margin, p_margin = ttest_ind(zip_a['Margin'], zip_b['Margin'], equal_var=False)
        
        zip_results.append({
            'ZipA': top_zipcodes[i],
            'ZipB': top_zipcodes[j],
            'ClaimFreq_pvalue': p_freq,
            'Margin_pvalue': p_margin
        })

zip_results_df = pd.DataFrame(zip_results)
zip_results_df
```
